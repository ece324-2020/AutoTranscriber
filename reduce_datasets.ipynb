```
import os
import pickle
import random
```

```
# User parameters

percent_to_remove = 0.15
data_path = '/content/drive/My Drive/ECE324/Project/MAESTRO/test' # path to data directory
save_path = '/content/drive/My Drive/ECE324/Project' # path to save list of data keeping to
save_file_name = 'data_kept_test.pkl'
```

```
# Get all data files

data_list = []

for file in os.listdir(os.path.join(data_path, 'data')):
  data_list.append(file)

print(len(data_list))
```

```
# Randomize order of data and select subset to remove

random.shuffle(data_list)
data_to_remove = data_list[:int(len(data_list)*percent_to_remove)]
data_to_keep = data_list[int(len(data_list)*percent_to_remove):]
```

```
# Save data being kept

with open(os.path.join(save_path, save_file_name), 'wb') as file:
  pickle.dump(data_to_keep, file)
```

```
# Check data was saved successfully

with open(os.path.join(save_path, save_file_name), 'rb') as file:
  test_data = pickle.load(file)

print(data_to_keep)
print(test_data)
```

```
# Remove data from both input data and labels

for file in data_to_remove:
  os.remove(os.path.join(data_path,'labels',file[:-4]+'.mid'))
  os.remove(os.path.join(data_path,'data',file))
```

```
# Check if successfully removed files

count = 0
for file in os.listdir(os.path.join(data_path, 'data')):
  count += 1
print('Number of data files', count)

count = 0
for file in os.listdir(os.path.join(data_path, 'labels')):
  count += 1
print('Number of label files', count)
```
